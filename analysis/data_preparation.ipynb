```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

```

```python
load_dotenv()

DB_NAME = os.getenv("MYSQL_DATABASE")
DB_USER = os.getenv("MYSQL_USER")
DB_PASSWORD = os.getenv("MYSQL_PASSWORD")
DB_HOST = "127.0.0.1"
DB_PORT = 3306
```

```python
engine = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

```python
df = pd.DataFrame({
    "order_id": [1, 2, 3],
    "product": ["Espresso", "Cappuccino", "Latte"],
    "quantity": [2, 1, 3],
    "price": [3.5, 4.0, 4.5]
})
```

```python
df.to_sql(
    name="coffee_sales",
    con=engine,
    if_exists="replace",
    index=False
)

print("✅ Data loaded from .env and written to MySQL")
```

```python
df = pd.read_sql(
    sql="SELECT * FROM coffee_sales",
    con=engine
)
```
